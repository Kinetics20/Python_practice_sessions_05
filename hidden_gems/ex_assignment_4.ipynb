```python
def three_oldest_club(dataset):
    sorted_clubs = sorted(dataset, key=lambda x: x['founded'])
    the_oldest_club = sorted_clubs[:3]
    current_year = 2026
    res = []

    for club in the_oldest_club:
        name = club['name']
        age = current_year - club['founded']
        res.append((name, age))

    return res


three_oldest_club(handball_elite)
```

```python
def three_oldest_club_dict(dataset):

    return {f'club name: {club['name']}': 2026 - club['founded'] for club in sorted(dataset, key=lambda x: x['founded'])[:3]}

three_oldest_club_dict(handball_elite)
```

```python
def name_city_1st(dataset):
    return {name['name']: name['city'] for name in dataset if name['successes_in_cl']['1st'] > 1}

name_city_1st(handball_elite)
```

```python
def name_city_sum_more_than_3(dataset):
    return {name['name']: name['city'] for name in dataset if sum(name['successes_in_cl'].values()) >= 4}

name_city_sum_more_than_3(handball_elite)
```

```python
def success_poland_team(dataset):
    return [(idx, name) for idx, name in enumerate([item for item in dataset
            if item['successes_in_cl']['1st'] > 0 and item['successes_in_cl']['2nd'] > 0 and item['successes_in_cl']['3rd'] > 0
            and item['country'] == 'Poland'
            ], 1)]

success_poland_team(handball_elite)
```

```python
def success_poland_team_2(dataset):
    return [item for item in dataset
            if any(item['successes_in_cl'][stage] for stage in ('1st', '2nd', '3rd'))
            and item['country'] == 'Poland'
            ]

success_poland_team_2(handball_elite)
```

```python
from collections import Counter

def amount_countries(dataset):
    return (Counter(item['country'] for item in dataset)).most_common(3)

amount_countries(handball_elite)
```
