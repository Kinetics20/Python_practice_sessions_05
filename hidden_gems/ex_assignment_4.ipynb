```python
def three_oldest_club(dataset):
    sorted_clubs = sorted(dataset, key=lambda x: x['founded'])
    the_oldest_club = sorted_clubs[:3]
    current_year = 2026
    res = []

    for club in the_oldest_club:
        name = club['name']
        age = current_year - club['founded']
        res.append((name, age))

    return res


three_oldest_club(handball_elite)
```

```python
def three_oldest_club_dict(dataset):
    return {f'club name: {club['name']}': 2026 - club['founded'] for club in
            sorted(dataset, key=lambda x: x['founded'])[:3]}


three_oldest_club_dict(handball_elite)
```

```python
from collections import Counter


def amount_countries(dataset):
    return (Counter(item['country'] for item in dataset)).most_common(3)


amount_countries(handball_elite)
```

```python
def amount_countries_for(dataset):
    countries_count = {}

    for club in dataset:
        country = club['country']
        if country in countries_count:
            countries_count[country] += 1
        else:
            countries_count[country] = 1

    return sorted(countries_count.items(), key=lambda x: x[-1], reverse=True)[:3]


amount_countries_for(handball_elite)
```

```python
def clubs_by_country(dataset, country):
    return [item['name'] for item in dataset if country in item['country']]

clubs_by_country(handball_elite, 'Poland')
```

```python
def clubs_founded_before(dataset, year):
    return [{club['name']: club['founded']} for club in dataset if club['founded'] < year]

clubs_founded_before(handball_elite, 1955)
```

```python
def clubs_with_color(dataset, color):
    return [{(item['name']): item['colors']} for item in dataset if color in item['colors']]

clubs_with_color(handball_elite, 'white')
```

```python
def count_clubs_by_country(dataset):
    clubs = {}

    for club in dataset:
        country = club['country']
        if country in clubs:
            clubs[country] += 1
        else:
            clubs[country] = 1
    return sorted(clubs.items(), key=lambda x: x[-1], reverse=True)

count_clubs_by_country(handball_elite)
```

```python
def oldest_and_youngest_club(dataset):
    clubs_year = {club['name']: 2026 - club['founded'] for club in sorted(dataset, key=lambda x: x['founded'])}

    return clubs_year

oldest_and_youngest_club(handball_elite)
```

```python
def oldest_and_youngest_club_2(dataset):
    clubs_year = {club['name']: 2026 - club['founded'] for club in sorted(dataset, key=lambda x: x['founded'])}
    names = [name for name in clubs_year.keys()]
    return names[0], names[-1]

oldest_and_youngest_club_2(handball_elite)
```

```python
def oldest_and_youngest_club_3(dataset):
    the_oldest_club = min(dataset, key=lambda x: x['founded'])
    the_youngest_club = max(dataset, key=lambda x: x['founded'])

    return the_oldest_club['name'], the_youngest_club['name']

oldest_and_youngest_club_3(handball_elite)
```

```python
def avg_capacity_by_country(dataset, country):
    capacity = []
    for club in dataset:
        if club['country'] == country:
            capacity.append(club['capacity'])

    return sum(capacity) / len(capacity)

avg_capacity_by_country(handball_elite, 'Poland')
```

```python
def avg_capacity_by_country_2(dataset, country):

    capacity_list = [club['capacity'] for club in dataset if club['country'] == country]
    return sum(capacity_list) / len(capacity_list) if capacity_list else 0

avg_capacity_by_country_2(handball_elite, 'Poland')
```
