```python
def avg_age_politicians_en_rus(dataset):
    list_of_ages = [person['age'] for person in dataset if
                    ('English' in person['language'] or 'Russian' in person['language'])]
    return round(sum(list_of_ages) / len(list_of_ages))


avg_age_politicians_en_rus(politics)
```
