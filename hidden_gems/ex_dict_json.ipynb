```python
import json
```

```python
json_txt = '{"name": "Adam", "age": 30, "skills": ["Python", "ML"]}'
data = json.loads(json_txt)
data
```

```python
data.get('salary', 1000)
# data
```

```python
data['salary'] = 1000
data
```

```python
data['skills'] += ['SQL']
data
```

```python
back_data = json.dumps(data, indent=2)
print(back_data)
```

```python
type(back_data)
```

```python
nd = json.loads(back_data)
nd
```

```python
with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(nd, f, indent=2)
```

```python
with open('data.json', 'r', encoding='utf-8') as f:
    json_txt = f.read()

print(json_txt)
```

```python
with open('data.json', 'r', encoding='utf-8') as f:
    data_dict = json.load(f)

data_dict
```
