```python
def simple_range(start: int, stop: int | None=None, step: int=1) ->list[int]:
    if step in (None, 0):
        raise ZeroDivisionError('Step must be not zero.')

    if stop is None:
        stop = start
        start = 0

    r_list: list[int] = []
    current: int = start

    while len(r_list) < (stop - start) / step:
        r_list.append(current)
        current += step

    return r_list
```

```python
def get_avg_age(dataset):
    # result = 0
    #
    # for item in dataset:
    #     result += item['age']
    #
    # return round(result / len(dataset))

    # return round(sum(x['age'] for x in dataset) / len(dataset))

    return round(sum(map(lambda x: x['age'], dataset)) / len(dataset))

get_avg_age(data)
```

```python
from operator import itemgetter

def get_youngest(dataset):
    age = dataset[0]['age']
    result = None

    for item in dataset:
        if item['age'] < age:
            age = item['age']
            result = item

    return result

    # return min(dataset, key=lambda x: x['age'])
    return min(dataset, key=itemgetter('age'))

get_youngest(data)
```
