```python
def sort_by(key: Callable[[list], Any], reverse: bool=False) -> Callable[[list], list]:
    return lambda lst: sorted(lst, reverse=reverse, key=key)


sort_numbers = sort_by(key=lambda x: x)

data = [5, 2, 9, 1, 7]
result = sort_numbers(data)

result
```
