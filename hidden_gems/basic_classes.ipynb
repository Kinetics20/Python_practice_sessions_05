```python
class Animal:
    def __init__(self, species: str, name: str, leg_count: int):
        self.species = species
        self.name = name
        self.leg_count = leg_count

    def change_name(self, new_name):
        self.name = new_name

    def __repr__(self):
        return (
            f"Animal(species={self.species!r}, "
            f"name={self.name!r}, "
            f"leg_count={self.leg_count!r})"
        )

    def __str__(self):
        return f"{self.name} the {self.species} ({self.leg_count} legs)"

    def __format__(self, format_spec: str):
        if format_spec == "" or format_spec == "default":
            return str(self)

        if format_spec == "short":
            return f"{self.name} ({self.leg_count} legs)"

        if format_spec == "long":
            return f"{self.name} is a {self.species} with {self.leg_count} legs."

        raise ValueError(f"Unknown format specifier: {format_spec!r}")

animals = [
    Animal('dog', 'Almo', 4),
    Animal('cat', 'Mike', 4),
    Animal('parrot', 'Henry', 2),
]

animals[0].change_name('Katty')
animals[0].name
```

```python
def print_animals(anim):
    print('---- SPECIES ----+------ NAME -------+---- LEGS -----')
    for x in anim:
        print(f'{x.species:^17} {x.name:^18} {x.leg_count:^18}')
print()
print_animals(animals)
```
