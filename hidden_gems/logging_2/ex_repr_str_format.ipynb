```python
class Car:
    def __init__(self, brand: str, model: str, year: int):
        self.brand = brand
        self.model = model
        self.year = year

    def __repr__(self):
        return(
            f'Car(brand={self.brand!r}, '
            f'model={self.model!r}, '
            f'year={self.year!r})'
        )

    def __str__(self):
        return f'Model {self.model} is a brand of {self.brand} from {self.year} year.'

    def __format__(self, format_spec: str):
        format_spec = format_spec.strip()
        if format_spec in ('', 'default'):
            return str(self)

        if format_spec == 'short':
            return f'The car is: {self.brand}, {self.model}'

        if format_spec == 'long':
            return f'It is : {self.brand}, {self.model} from {self.year} year.'

        raise ValueError(f'Unknown format specifier: {format_spec!r}')
```

```python
a = Car('Fiat', 'Uno', 1999)
```

```python
str(a)
```
