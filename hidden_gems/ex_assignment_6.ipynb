#### 1.Zwróć listę tytułów seriali, które: są z gatunku "crime" lub "thriller", mają co najmniej 4 sezony, oraz mają emmy_wins ≥ 5.
#### 2.Zwróć listę tytułów seriali, które: pochodzą z "UK", mają liczbę odcinków ≥ 30,oraz mają w awards.other wpis zawierający słowo "BAFTA".
#### 3. Zwróć listę krotek (title, award_tier) tylko dla seriali, gdzie:
#### - award_tier = "top" jeśli awards.emmy_wins ≥ 10,
#### - award_tier = "mid" jeśli awards.emmy_wins jest w przedziale 1–9,
#### - award_tier = "none" jeśli awards.emmy_wins = 0 lub brak tego pola.

```python
def crime_or_thriller(dataset):
    return [title['title'] for title in dataset
            if ('crime' in title['genre'] or 'thriller' in title['genre'])
            and title['seasons'] >= 4
            and title['awards'].get('emmy_wins', 0) >= 5
            ]


crime_or_thriller(tv_series)
```

```python
def uk_series(dataset):
    return [title['title'] for title in dataset
            if "UK" in title['origin']
            and title['episodes'] >= 30
            and any('BAFTA' in award for award in title['awards']['other'])
            ]

uk_series(tv_series)
```
