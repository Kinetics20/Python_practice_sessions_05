```python
# 3
def amount_of_features(dataset):
    return sum([len(item['characteristic features']) for item in dataset
                if 'Bartosz Walaszek' in item['director']])


amount_of_features(series)
```

```python
def amount_of_features_2(dataset):
    set_of_features = {len(item['characteristic features']) for item in dataset
                       if 'Bartosz Walaszek' in item['director']}

    return sum([item for item in set_of_features])


amount_of_features_2(series)
```

```python
def title_avg_air(dataset):
    return [{item['title']: round(item['episodes'] / (max(item['air date']) - min(item['air date'])))} for item in dataset]

title_avg_air(series)
```

```python
def title_avg_air_2(dataset):
    return {item['title']: round(item['episodes'] / (max(item['air date']) - min(item['air date'])), 2) for item in dataset}

title_avg_air_2(series)
```

```python
def title_avg_air_key(dataset):
    title_avg = {item['title']: round(item['episodes'] / (max(item['air date']) - min(item['air date'])), 2)
                 for item in dataset}

    return dict(sorted(title_avg.items(), key=lambda x: x[0], reverse=True))

title_avg_air_key(series)
```

```python
# 1
def title_avg_air_value(dataset):
    title_avg = {item['title']: round(item['episodes'] / (max(item['air date']) - min(item['air date'])), 2)
                 for item in dataset}

    return dict(sorted(title_avg.items(), key=lambda x: x[1], reverse=True))

title_avg_air_value(series)
```
