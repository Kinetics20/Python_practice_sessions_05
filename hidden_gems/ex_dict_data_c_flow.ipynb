```python
def count_each_hobbies(items):
    res = {}

    for item in items:
        for hobby in item['hobbies']:
            if hobby in res:
                res[hobby] += 1
            else:
                res[hobby] = 1
    return res


count_each_hobbies(data)
```

```python
from collections import Counter


def count_each_hobbies_counter(items):
    res = []

    for item in items:
        res.extend(item['hobbies'])

    return Counter([x for x in res])


count_each_hobbies_counter(data)
```

```python
from collections import Counter


def count_each_hobbies_counter_flatten(items):
    return Counter(x for y in items for x in y['hobbies'])


count_each_hobbies_counter_flatten(data)
```

```python
from collections import Counter


def count_each_hobbies_counter_pf(items):
    return Counter(reduce(lambda acc, ce: acc + ce['hobbies'], items, []))


count_each_hobbies_counter_pf(data)
```

```python
def two_elements_list_of_str(dataset):
    res_join = [
        ''.join(
            str(v)[:2] for v in list(x.values())[:-1]
        ) + ''.join(
            str(h)[:2] for h in list(x.values())[-1]
        )
        for x in dataset
    ]

    return res_join


def res_join_hashed(list_of_str):
    return [hash(x) for x in list_of_str]


def res_reduce(dataset):
    return reduce(lambda acc, ce: ce(acc), [two_elements_list_of_str, res_join_hashed], dataset)


res_reduce(data)
```

```python
def join_head(d: dict) -> str:
    return ''.join(str(v)[:2] for v in list(d.values())[:-1])


def join_tail(d: dict) -> str:
    return ''.join(str(h)[:2] for h in list(d.values())[-1])


def join_all(d: dict) -> str:
    return join_head(d) + join_tail(d)


def map_join(dataset: list[dict]) -> list[str]:
    return [join_all(x) for x in dataset]


def hash_all(items: list[str]) -> list[int]:
    return [hash(x) for x in items]


def map_join(dataset: list[dict]) -> list[str]:
    return [join_all(x) for x in dataset]


def hash_all(items: list[str]) -> list[int]:
    return [hash(x) for x in items]


def res_reduce(dataset):
    return reduce(
        lambda acc, f: f(acc),
        [
            map_join,
            hash_all
        ],
        dataset
    )


res_reduce(data)
```

```python
def two_elements_list_of_str(dataset):
    return map(
        lambda x: ''.join(str(v)[:2] for v in list(x.values())[:-1]) +
                  ''.join(str(h)[:2] for h in list(x.values())[-1]),
        dataset
    )


def res_join_hashed(iterable):
    return map(hash, iterable)


def res_reduce(dataset):
    return reduce(
        lambda acc, fn: fn(acc),
        [two_elements_list_of_str, res_join_hashed],
        dataset
    )


result = list(res_reduce(data))
result
```
