```python
data = [
    {
        "name": "pawel",
        "age": 40,
        "city": "krakow",
        "hobbies": ["python", "js", "drugs"]
    },
    {
        'name': 'pete',
        'age': 50,
        'city': 'warsaw',
        'hobbies': ['python', 'snorkeling', 'films']
    },
    {
        'name': 'katarzyna',
        'age': 32,
        'city': 'krakow',
        'hobbies': ['ksiazki', 'plywanie', 'seriale']
    },
    {
        "name": "agata",
        "age": 28,
        "city": "lipnicawielka",
        "hobbies": ["hiking", "crocheting", "reading"]
    },
    {
        "name": "przemek",
        "age": 40,
        "city": "krakow",
        "hobbies": ["książki", "place zabaw", "muzyka"]
    },
    {'name': 'Amos', 'age': 45, 'city': 'Olkusz', 'hobbies': ['cycling', 'football', 'chess']},
    {
        "name": "andrzej",
        "age": 29,
        "city": "katowice",
        "hobbies": ["3dprinting", "history", "skiing"]
    }
]
```

```python
def unique_hobby(dataset):
    res = set()

    for item in dataset:
        res |= set(item['hobbies'])

    return list(enumerate(res))


unique_hobby(data)
```

```python
def unique_hobby_for(dataset):
    e_set = set()

    for people in dataset:
        for hobby in people['hobbies']:
            e_set.add(hobby)

    return e_set


unique_hobby_for(data)
```

```python
def unique_hobby_for_flatten(dataset):
    return len({hobby for people in dataset for hobby in people['hobbies']})


unique_hobby_for_flatten(data)
```

```python
def unique_hobby_red(dataset):
    return reduce(lambda acc, ce: acc | set(ce['hobbies']), dataset, set())


unique_hobby_red(data)
```

```python
def count_unique_hobby(dataset):
    h_dict = {}

    for people in dataset:
        for hobby in people['hobbies']:
            if hobby in h_dict:
                h_dict[hobby] += 1
            else:
                h_dict[hobby] = 1

    return h_dict


count_unique_hobby(data)
```

```python
def count_each_hobbies_counter(dataset):
    e_list = []

    for people in dataset:
        e_list.extend(people['hobbies'])

    return Counter([x for x in e_list])


count_each_hobbies_counter(data)
```

```python
def count_each_hobbies_counter_flatten(dataset):
    return Counter([hobby for people in dataset for hobby in people['hobbies']])


count_each_hobbies_counter_flatten(data)
```

```python
def count_each_hobbies_counter_reduce(dataset):
    return Counter(reduce(lambda acc, ce: acc + ce['hobbies'], dataset, []))


count_each_hobbies_counter_reduce(data)
```

```python
def two_elem_hash(dataset):
    l_1 = [''.join(str(d)[:2] for d in list(x.values())[:-1]) for x in dataset]
    l_2 = [''.join(str(d)[:2] for d in list(x.values())[-1]) for x in dataset]

    all_l = list(map(lambda x, y: x + y, l_1, l_2))
    # all_l = [a + b for a, b in zip(l_1, l_2)]

    return [hash(x) for x in all_l]


two_elem_hash(data)
```

```python
def two_elem_hash_for(dataset):
    a_list = []

    for x in dataset:
        values = list(x.values())

        part_1 = ''.join(str(z)[:2] for z in values[:-1])
        part_2 = ''.join(str(z)[:2] for z in values[-1])

        all_added = part_1 + part_2

        a_list.append(hash(all_added))

    return a_list

two_elem_hash_for(data)
```
