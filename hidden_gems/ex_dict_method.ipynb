```python
def draw_triangle(n):
    output = []
    for i in range(n):
        txt = '*' * (i + 1)
        output.append(txt)
    return output

t = draw_triangle(10)
print('\n'.join(t))
```

```python
import time
import sys

v = 1
i = 0
print('')

while True:
    sys.stdout.write('\r' + (' ' * i) + 'Break until next week')
    sys.stdout.flush()

    i += v
    if i < 1:
        v = 1
    elif i > 180:
        v = -1
    time.sleep(0.1)
```

```python
def sum_all(*numbers):
    return sum(numbers)

print(sum_all(1, 2, 3, 4))
```

```python
def print_info(**kwargs):
    for k, v in kwargs.items():
        print(f"{k}: {v}")

print_info(name="Alice", age=30, city="NYC")
```

```python
def complex_func(a, b, *args, key1=None, **kwargs):
    print(f"a={a}, b={b}")
    print(f"args={args}")
    print(f"key1={key1}")
    print(f"kwargs={kwargs}")

print(complex_func(1, 2, 3, 4, 5, key1="value", extra="data"))
```
