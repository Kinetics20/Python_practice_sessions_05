```python
users = {
    "user1": {"name": "Julia", "age": 25},
    "user2": {"name": "John", "age": 30}
}
```

```python
parameters = {'n_clusters': [2, 3, 4, 5, 10, 20, 30]}
```

```python
[x for x in parameters.values()]
```

```python
[{'n_clusters': x} for x in parameters.values()]
```

```python
[[{'n_clusters': y} for y in x] for x in parameters.values()]
```

```python
[[{k: y} for y in x] for k, x in parameters.items()]
```
