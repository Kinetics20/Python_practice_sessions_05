##### Question 1: Find all electric or hybrid cars under $50,000 with average rating >= 4.3, return a list of tuples containing (model, manufacturer, price_usd, average_rating) sorted by average_rating in descending order.
##### Question 2: Calculate average emission rating grouped by fuel_type return a dictionary with fuel_type as key and average emission as value then find which fuel type has the lowest average emission rating.
##### Question 3: For all cars with horsepower > 220, create tuples of (model, manufacturer, price_per_hp_ratio) where price_per_hp_ratio = price_usd / horsepower return only cars with price_per_hp_ratio < 220, sorted by price_per_hp_ratio ascending.

```python
def cars_price_per_hp_ratio(dataset):
    res = []

    for car in dataset:
        price_per_hp_ratio = round(car['price_usd'] / car['horsepower'], 3)
        if (car['horsepower'] > 220) and (price_per_hp_ratio < 220):
            res.append((car['model'], car['manufacturer'], price_per_hp_ratio))

    return sorted(res, key=lambda x: x[-1])

cars_price_per_hp_ratio(car_manufacturers)
```

```python
def electric_or_hybrid(dataset):
    res = []

    for car in dataset:
        avg = sum(car['ratings']) / len(car['ratings'])
        if car['fuel_type'] == 'electric' or car['fuel_type'] == 'hybrid':
            res.append((car['model'], car['manufacturer'], car['price_usd'], avg))

    return sorted(res, key=lambda x: x[-1], reverse=True)


electric_or_hybrid(car_manufacturers)
```

```python
def avg_emission(dataset):
    avg = {}

    for fuel_type in ['electric', 'hybrid', 'petrol']:
        emissions = [car['emission_rating_g_km'] for car in dataset
                     if car['fuel_type'] == fuel_type]

        if emissions:
            avg[fuel_type] = sum(emissions) / len(emissions)

    return min(avg.keys(), key=lambda item: item[-1])


avg_emission(car_manufacturers)
```
