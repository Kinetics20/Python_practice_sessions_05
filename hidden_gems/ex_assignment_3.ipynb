#### Zwróć listę nazw tras, które:
#### 1.mają tag "views", nie są o trudności "easy", mają przewyższenie co najmniej 800 m.
#### 2.nazwy tras, które: są dostępne wiosną i latem (oba sezony), mają średnią ocen ≥ 4.5, mają dystans krótszy niż 40 km
#### 3. Zwróć listę krotek (name, popularity) tylko dla tras dostępnych latem, gdzi popularity = "popular" jeśli liczba ocen ≥ 4

```python
def route_names_tag_views(dataset):
    route_names = []

    for route in dataset:
        if 'views' in route['tags'] and route['difficulty'] != 'easy' and route['elevation_gain_m'] >= 800:
            route_names.append(route['name'])

    return route_names


route_names_tag_views(bike_routes)
```

```python
def route_names_spring_summer(dataset):
    route_names = []

    for route in dataset:
        avg_score = sum(route['ratings']) / len(route['ratings'])

        if 'spring' in route['seasons'] and 'summer' in route['seasons'] and avg_score >= 4.5 and route[
            'distance_km'] < 40:
            route_names.append(route['name'])

    return route_names


route_names_spring_summer(bike_routes)
```

```python
def route_tuple_name_popularity(dataset):
    route_names = []

    for route in dataset:
        popular = len(route['ratings'])

        if (len(route['seasons']) == 1 and 'summer' in route['seasons']) and popular >= 4:
            route_names.append((route['name'], 'popularity'))

    return route_names


route_tuple_name_popularity(bike_routes)
```

```python
def route_tuple_name_popularity_avg(dataset):
    route_names = []

    for route in dataset:
        avg_score = sum(route['ratings']) / len(route['ratings'])

        if (len(route['seasons']) == 1 and 'summer' in route['seasons']) and avg_score >= 4:
            intensity_level = 'high' if route['elevation_gain_m'] > 1000 else ('medium' if route['elevation_gain_m'] > 500 else 'easy')
            route_names.append((route['name'], 'popularity', f'intensity_level: {intensity_level}'))

    return route_names

route_tuple_name_popularity_avg(bike_routes)
```
