```python
def gen_id(id_ = 1):

    def next_id(new_id = None):

        if new_id is not None:
            next_id.id_ = new_id

        result = next_id.id_
        next_id.id_ += 1
        return result

    next_id.id_ = id_
    return next_id

next_gen = gen_id(42)
print(next_gen())
```

```python
print(next_gen())
```

```python
print(next_gen(888))
```

```python
from collections.abc import Callable


def power_factory(exp: int) -> Callable[[int], int]:
    def power(x: int) -> int:
        return x ** exp
    return power
```

```python
power4 = power_factory(4)
power5 = power_factory(5)
print(power4(4), power5(5))
```

```python
def min_length_validator(min_length: int) -> Callable[[str], bool]:
    def validate(value: str) -> bool:
        return len(value) >= min_length
    return validate

is_strong_pass = min_length_validator(14)
is_strong_pass('TopSecret123')
```

```python
def memoize(fn: Callable[[int, int], int]) -> Callable[[int], int]:
    cache: dict[int, int] = {}

    def inner(x: int) -> int:
        if x not in cache:
            cache[x] = fn(x)
        return cache[x]

    return inner

@memoize
def square(x: int) -> int:
    print('computing...')
    return x * x

square_cached = memoize(square)
square_cached(10)
```
