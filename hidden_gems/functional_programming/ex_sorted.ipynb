```python
any_list = [-5, -20, 19, -17, -8, 89, -59, 19, 0, 693, -542, 588, 903, -472, -16, 2, 663, -21, 47, 37, -10, -731, -5, -366, -85, 174, -89, 99, -288, 26, -7, -867, -586, -351, 92, -48]
```

```python
def men_from_boys(y):
    return sorted(set(y), key=lambda x: (x % 2, x if not x % 2 else -x))
```

```python
men_from_boys(any_list)
```
