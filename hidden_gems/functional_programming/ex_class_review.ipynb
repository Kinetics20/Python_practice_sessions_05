```python
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        return f'{self.name} says woof!'

d1 = Dog('Zoya')
d1.bark()
```

```python
class UserSth:
    count = 0

    def __init__(self):
        UserSth.count += 1

    @classmethod
    def how_many(cls):
        return cls.count

us1 = UserSth()
us2 = UserSth()
us3 = UserSth()
UserSth.how_many()
```

```python
class UserTwo:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def from_str(cls, data):
        name, age = data.split(', ')
        return cls(name, int(age))

    def __repr__(self):
        return f'UserTwo(name={self.name!r}, age={self.age})'

ut1 = UserTwo.from_str('Mike, 45')
ut1.name, ut1.age
```
