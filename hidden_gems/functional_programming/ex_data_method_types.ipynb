```python
def simple_calc(a, b, operator):
    dict_oper = {
        '+': lambda: a + b,
        '-': lambda: a - b,
        '*': lambda: a + b,
        '/': lambda: a / b
    }

    return dict_oper[operator]()
```

```python
simple_calc(10, 20, '+')
```

```python
def simple_calc_match(a, b, operator):
    match operator:
        case '+':
            return a + b
        case '-':
            return a - b
        case _:
            raise ValueError('Unknown operator')
```

```python
simple_calc_match(10, 20, '+')
```

```python
def memo(fn):
    cache = {}

    def inner(a, b):
        if f'{a}{b}' not in cache:
            cache[f'{a}{b}'] = fn(a, b)
        print(cache)
        return cache[f'{a}{b}']

    return inner

@memo
def add(a, b):
    return a + b

@memo
def multiply(a, b):
    return a * b
```

```python
a = add(1, 2)
b = multiply(3, 4)
print(a, b)
```

```python
def divide(a, b):
    return a / b

divide_m = memo(divide)
divide_m(20, 5)
divide_m(40, 1)
divide_m(90, 1)
```
