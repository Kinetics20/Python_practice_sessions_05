```python
d = list(range(15))
```

```python
f_odd = next(
    filter(lambda x: x % 2, d), None
)
f_odd
```

```python
z = next(
    (x for x in d if x == 20), None
)
print(z)
```

```python
try:
    num = int(input('Enter a number: '))
except ValueError:
    print('It is not a number')
```

```python
try:
    num = int(input('Enter a number: '))

    if num < 0:
        print('It is negative')
    else:
        print('It is positive')
except ValueError:
    print('It is not a number')
```

```python
try:
    a = int(input('a: '))
    b = int(input('b: '))
    print(a / b)

except ValueError:
    print('Invalid input')
except ZeroDivisionError:
    print('Division by zero!')
```

```python
try:
    age = int(input("Enter the age: "))

    if age < 0:
        raise ValueError("The age cannot be negative!")

except ValueError as e:
    print(f"Error: {e}")
```

```python
while True:
    try:
        number = int(input("Enter a number: "))
    except ValueError:
        print("Invalid number, try again.")
    else:
        print(f"You entered: {number}")
        break
```
