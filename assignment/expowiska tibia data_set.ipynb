### questions :
1. Wiedząc, że nie poradzę sobie z potworami, które mają powyżej 600hp gdzie zdobędę najwięcej doświadczenia (exp) ?
2. Który z potworów ma najlepszy współczynnik expa do hp (im więcej expa za mniejszą ilość hp tym lepiej)
3. Wiedząc, że z każdego potwora wypada zawsze cały loot, na którym expowisku najwięcej zarobie?

- potwory na expowiskach pojawią się tylko raz i po zabiciu nie pojawiają się ponownie

```python
def max_exp(dataset):
    max_exp_value = max(monster['exp'] for location in dataset for monster in location['monsters'])
    
    return max_exp_value
    
    # return [monster['name'] for location in dataset for monster in location['monsters'] if monster['hp'] < 600 and monster['exp'] >= max_exp_value]
```

```python
def best_location_for_exp(dataset):
    max_exp_location = None
    max_exp_total = 0
    
    for location in dataset:
        
        total_exp = sum(monster['exp'] for monster in location['monsters'] if monster['hp'] <= 600)
        
        
        if total_exp > max_exp_total:
            max_exp_total = total_exp
            max_exp_location = location['name']
    
    return {'location': max_exp_location, 'total_exp': max_exp_total}
```

```python
best_location_for_exp(expowiska)
```

```python
def best_exp_to_hp_ratio(dataset):
    best_monster = None
    best_ratio = 0
    
    for location in dataset:
        for monster in location['monsters']:
            
            ratio = monster['exp'] / monster['hp']
            
           
            if ratio > best_ratio:
                best_ratio = ratio
                best_monster = {
                    'monster': monster['name'],
                    'exp_to_hp': best_ratio
                }
    
    return best_monster
```

```python
best_exp_to_hp_ratio(expowiska)
```

```python
def best_loot_location(dataset):
    max_loot_value = 0
    best_location = None
    
    for location in dataset:
        total_loot_value = 0
        for monster in location['monsters']:
            
            loot_value_per_monster = sum(item['npc_value'] for item in monster['loot'])
            total_loot_value += loot_value_per_monster * monster['count_monster_on_spawn']
        
       
        if total_loot_value > max_loot_value:
            max_loot_value = total_loot_value
            best_location = {
                'location': location['name'],
                'total_loot_value': max_loot_value
            }
    
    return best_location
```
